# Aula 02

Para essa aula usaremos o pacote **readr** que possui a métodos para leituras de diferentes tipos de arquivos. O método **read_csv()** será o responsável pela letura do dataset _duracao.csv_que agrega dados sobre quantos dias cada usuário levou para terminar cada curso. Iremos atribuir o dataset a variável 'duracao'

```r
install.packages('readr')
```

```r
library('readr')
```

```r
duracao <- read_csv('duracao.csv')
```

```r
View(duracao)
```

Como os nomes das variáveis estão de entendimento complicado, é interessante que mudeos os nomes. A função **rename()** será a responsável por essa mudança. Essa função recebe o dataset como primeiro parâmetro e um *conjunto* como segundo parâmetro (*replace*). Esse conjunto no R é marcado pela função **c()**, que recebe como parâmetros os pares que irão compor a troca dos nomes

```r
duracao <- rename(duracao, replace = c('user_id' = 'aluno', 'course_id' = 'curso', 'timeToFinish' = 'dias'))
```

```r
View(duracao)
```

Para melhor visualização e exploração dos dados, trabalhar com gráficos é a melhor opção. Nativaente, o R possui o método **plot()** que recebe como parâmetro o conjunto de dados que se quer plotar. Por patrão, a função plot utiliza um gráfico de pontos.

```r
plot(duracao$dias)
```

Como os dados analisados são melhor explorados em um histograma, o método **hist()** será o nosso escolhido para plotar esse gráfico.

```r
hist(duracao$dias)
```

O parâmetro *breaks* representa o número de classes que se quer quebrar o histograma.

```r
hist(duracao$dias, breaks = 3)
```

```r
hist(duracao$dias, breaks = 20)
```

Adicionando labels no gráficos alterando os valores dos parâmetros, *main* (título do gráfico), *ylab* e *xlab* (legendas dos eixos)

```r
hist(duracao$dias, breaks = 20, main = 'Histograma do Tempo', ylab = 'Quantidade', xlab = 'Tempo')
```

O parâmetro *ylim = c(0,2500)* determina os limites dos eixos do gráfico. no caso o eixo y vai de 0 até 2500.

```r
hist(duracao$dias, breaks = 20, main = 'Histograma do Tempo', ylab = 'Quantidade', xlab = 'Tempo', ylim = c(0,2500))
```

O parâmetro col adiciona cores ao gráfico. esse parâmetro pode adicionar cores nomeadas ou cores em hash (#)

```r
hist(duracao$dias, breaks = 20, main = 'Histograma do Tempo', ylab = 'Quantidade', xlab = 'Tempo', ylim = c(0,2500), col = 'blue')
```

a função **jpeg()** diz para o R que uma figura vai ser criada. a função hist() é responsável por criar o gráfico em si e a função **dev.off()** é a responsável por fechar a figura

```r
jpeg('histograma.jpg')
hist(duracao$dias, breaks = 20, main = 'Histograma do Tempo', ylab = 'Quantidade', xlab = 'Tempo', ylim = c(0,2500), col = 'blue')
dev.off()
```

# Aula 03

Começaremos com as análises estatísticas dos dados. a primeira a ser realizada é a média com a função **mean()**. No primeiro exemplo, o conjunto de dados possui dados que não são numéricos, logo a média não pode ser calculada.

```r
mean(duracao$dias)
```

Para remover os dados que não são numéricos, usamos o parâmetro **na.rm = TRUE **para excluir do cálculos os valores não-numéricos.

```r
mean(duracao$dias, na.rm = TRUE)
```

A mediana é a próxima medida descritiva dos dados que será utilizada; o mesmo erro anterior é encontrado...

```r
 median(duracao$dias)
```

... e a mesma solução é aplicada.

```r
 median(duracao$dias, na.rm = T)
```

```r
summary(duracao$dias)
```

Na tabela, o valor do primeiro (1st Qu.) é 2.00, indicando que em 25% dos casos os alunos levaram menos de 2 dias para concluir o curso. Já o valor do terceiro (3rd Qu.) indica que 25% dos alunos levaram mais de 45 dias para concluir os cursos. Por fim, em Min. temos o valor mínimo 0.00, que já conhecíamos, pois não é possível um aluno levar menos que 0 dias para concluir um curso. E o valor de dados faltantes (NA's), 3828, indica que na amostra do banco de dados não há informações sobre a quantidade de dias que 3828 alunos levaram para concluir um curso.

Refraseando: 3828 alunos desistiram e não concluíram os cursos ou ainda não tinham concluído os cursos quando o banco de dados foi gerado. Agora, se 3828 alunos não concluíram ou concluíram os cursos após o envio da amostra do banco de dados, qual a proporção disso em relação ao total de alunos? Se isso for um problema, qual o seu tamanho? Para calcularmos, podemos usar funcionalidades do RStudio, que também funciona como calculadora, realizando diversas contas.

Para o cálculo dessa proporção, primeiro precisaremos descobrir o número de alunos do banco de dados por meio da dimensão **(dim)**:

```r
dim(duracao)[1]
```

O número 6366 representa o tamanho da amostra, o número de matrículas. Poderemos ter, por exemplo, um mesmo aluno sendo contabilizado mais de uma vez na mesma amostra. Certamente temos o mesmo curso repetidas vezes, também. Em 6366, há a junção de número de matrículas, e o dado individual indica um aluno fazendo um determinado curso. Precisaremos dividir o número de dados indisponíveis pelo número total. Podemos calcular isto diretamente no R Script:



```r
3828/6366
```

O resultado está em decimais (0.6013195), e significa que 60.13% das pessoas matriculadas não concluíram os cursos por desistência, ou concluíram após o envio da amostra. Como alguns cursos levam tempo até a conclusão, é natural depararmos com casos como esses.Se necessário, levamos essa informação para a empresa, que saberá melhor como lidar com ela.

```r
(3828/6366) * 100
```

Mas a que se refere a porcentagem 60.13%? Não sabemos as dimensões da amostra, portanto vamos colocar esse dado em comparação com o todo com que estamos trabalhando. Calcularemos a quantidade de cursos únicos para passarmos uma informação completa para a empresa, por meio de length. Como queremos calcular somente os casos únicos, utilizaremos unique:

```r
length(unique(duracao$curso))
```

Ou seja, na amostra há 264 cursos únicos. Por fim, faremos a mesma coisa para definirmos a quantidade de alunos únicos:

```r
length(unique(duracao$aluno))
```

Poderíamos passar a informação que obtivemos para a empresa, da seguinte forma:

"Na amostra que vocês nos deram, há 6366 matrículas, 264 cursos e 484 alunos únicos, dos quais 60.13% não concluíram os cursos por desistência, ou concluíram somente após o envio da amostra do banco de dados."

```r
popularidade <- read_csv('popularidade.csv')
```

```r
sumario_estatistico <- aggregate(duracao$dias, list(duracao$curso), mean, na.rm = T)
```

```r
head(sumario_estatistico)
```

```r
populariade_e_duracao <- merge(sumario_estatistico, popularidade, br = 'curso')
```

```r
sumario_estatistico <- rename(sumario_estatistico, replace = c('Group.1'='curso', 'x'='dias'))
```

```r
head(sumario_estatistico)
```

```r
popularidade <- rename(popularidade, c('course_id'='curso', 'freq'='popularidade'))
```

```r
head(popularidade)
```

```r
popularidade_e_duracao2 <- merge(sumario_estatistico, popularidade, by = 'curso')
```

```r
head(popularidade_e_duracao2)
```
